```python
#
duration = 50*60

# loop for every second, and draw a random number to see if there's a reward
n_rews = []
hrates = []
for k in range(100):
    thresh = 0.00825
    n_rewards = 0
    n_trials = 0
    trial_time = 0
    for k in range(duration):
        if np.random.rand()<thresh:
            n_rewards+=1
            n_trials+=1
        trial_time+=1

        if trial_time>=30:
            trial_time = 0
            n_trials+=1

    #print ("n rewards: ", n_rewards, " n_trials: ", n_trials)
    #print ("hit rate = ", n_rewards/n_trials)

    #
    n_rews.append(n_rewards)
    hrates.append(n_rewards/n_trials)

print ("mean hit rate: ", np.mean(hrates), " n rews: ", np.mean(n_rews))
```
